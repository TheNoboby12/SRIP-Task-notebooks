# Importing All Libraries.

```python
# import numpy as np
import jax.numpy as jnp
import matplotlib.pyplot as plt
from matplotlib import cm
import jax
```

#Bivariate Normal Distribution function


```python
def np_bivariate_normal_pdf(domain, mean, variance):
  X = jnp.arange(-domain+mean, domain+mean, variance)
  Y = jnp.arange(-domain+mean, domain+mean, variance)
  X, Y = jnp.meshgrid(X, Y)
  R = jnp.sqrt(X**2 + Y**2)
  Z = ((1. / jnp.sqrt(2 * jnp.pi)) * jnp.exp(-.5*R**2))
  return X+mean, Y+mean, Z
```

# Plotting the Bivariate Normal Ditribution Function

```python
def plt_plot_bivariate_normal_pdf(x, y, z):
  fig = plt.figure(figsize=(12, 6))
  ax = fig.gca(projection='3d')
  ax.plot_surface(x, y, z, 
                  cmap=cm.coolwarm,
                  linewidth=0, 
                  antialiased=True)
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel('z');
  plt.show()
```

# Sliders for Mean and Varience value using ipywidgets Library

```python
import ipywidgets as widgets

mean = widgets.FloatSlider(
    value=7.5,
    min=0,
    max=10.0,
    step=0.1,
    description='Mean:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)
var = widgets.FloatSlider(
    value=7.5,
    min=0.1,
    max=2,
    step=0.1,
    description='Variance:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)
mean
```
